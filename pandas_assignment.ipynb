```python
##Assignment 2: Working with Pandas

import pandas as pd

# 1. Create DataFrame [cite: 78, 79, 81, 84]
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}
df = pd.DataFrame(data)

# a. First five rows [cite: 85]
print(df.head())

# b. Summary statistics [cite: 86]
print(df[['Age', 'Salary']].describe())

# c. Average salary in HR [cite: 87]
avg_hr_salary = df[df['Department'] == 'HR']['Salary'].mean()
print(f"\nAverage HR Salary: {avg_hr_salary}")

# 2. Add Bonus column (10%) [cite: 88]
df['Bonus'] = df['Salary'] * 0.10

# 3. Filter Age 25-30 [cite: 89]
filtered_df = df[(df['Age'] >= 25) & (df['Age'] <= 30)]
print("\nEmployees aged 25-30:\n", filtered_df)

# 4. Group by Department [cite: 90]
avg_salary_dept = df.groupby('Department')['Salary'].mean()
print("\nAvg Salary by Dept:\n", avg_salary_dept)

# 5. Sort and Save [cite: 91]
df_sorted = df.sort_values(by='Salary', ascending=True)
df_sorted.to_csv('sorted_employees.csv', index=False)
```
