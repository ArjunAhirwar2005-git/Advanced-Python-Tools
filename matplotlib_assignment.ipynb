```python
## Assignment 3: Working with Matplotlib

import matplotlib.pyplot as plt
import numpy as np

# 1. Line Plot [cite: 98, 99, 100]
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
plt.figure()
plt.plot(x, y, marker='o')
plt.title("Simple Line Plot") # [cite: 102]
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.show()

# 2. Bar Graph [cite: 103, 104, 105]
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]
plt.figure()
plt.bar(students, marks, color=['blue', 'green', 'red', 'purple']) # [cite: 107]
plt.title("Student Marks")
plt.show()

# 3. Pie Chart [cite: 108, 109]
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
explode = (0.1, 0, 0, 0) # Highlighting highest revenue [cite: 111]
plt.figure()
plt.pie(revenue, labels=regions, autopct='%1.1f%%', explode=explode) # [cite: 110]
plt.title("Company Revenue by Region")
plt.show()

# 4. Histogram [cite: 112, 113]
data = np.random.randint(1, 101, 1000)
plt.figure()
plt.hist(data, bins=20, edgecolor='black')
plt.title("Frequency Distribution")
plt.show()
```
